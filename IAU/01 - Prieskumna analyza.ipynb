Ako bude ukázane nižšie, TT4 a TT4 measured(a im podobné) spolu súvisia. Chýbajúce hodnoty v týchto atribútoch môžeme doplniť pri týchto dvojiciach takým štýlom, že keď je napr pri TT4 hodnota NaN, tak meranie neprebehlo a TT4 measured je false.
Samozrejme je zopár výnimok, keď je meranie NaN a measured hodnota true, takže measured zmenime z 't' na 'f'. Obdobne je to aj s ďalšími, napr. pri FTI niesú pri false hodnotách NaN, ale '?'. Všetky hodnoty NaN zameníme na 0, kedže to je numericky atribút.

```python
df['T4U'].isnull().sum()
```

```python
df['T4U measured'].isnull().sum()
```

Tu sú všetky atribúty, ktoré majú v sebe 't'/'f' hodnoty reprezentujúce true/false. Vidíme, že ich formát nieje zjednotený. Zjednotíme hodnoty na 't' a 'f', pričom budú zmenené hodnoty 'TRUE','True','true' a tie, ktoré začínajú na 't.'. Obdobne 'FALSE','False','false' a tie ktoré začínajú 'f.', kedže vidíme, že sú tam za bodkou rôzne čísla.

```python
s = (df.eq('t') | df.eq('f')).any()
pole = s[s].index.values
pole
```
